# Tutorial Data Visualization

## Lollipop Charts

**A lollipop chart is basically a barplot, where the bar is transformed in a line and a dot. It shows the relationship between a numeric and a categoric variable.**

The lollipop plot is used exactly in the same situation than a barplot, but it is somewhat more appealing and convey as well the information. It is especially useful when you have several bars of the same height or if you are dealing with a large number of values and when the values are all high. Using a bar chart then can be visually aggressive, while a lollipop chart avoids to have a cluttered figure:

<img src="Pics/Bar_vs_Lollipop.png" height="600" width="600"/>

The lollipop chart also has some less fortunate features:

The center of the circle at the end of the lollipop marks the value, but the location of the center is difficult to judge. Furthermore, half of the circle extends beyond the value that it represents, making it slightly inaccurate.

## Airbnb Data

***Loading the Data***

The dataset here is about Airbnb data for Berlin. It's ready to use as I already cleaned it in a [previous analysis](https://github.com/BrittaInData/dsc-5-capstone-project-online-ds-ft-100118/blob/master/02_Notebook_2.ipynb). 

```python
# import the usual suspects ...
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
airbnb = pd.read_csv('Data/cleaned_airbnb')
print('Records:', airbnb.shape[0], '\nColumns:', airbnb.shape[1])
airbnb.head()
```

We have a `listing_id` for every accomodation, have information on its location, it costs, whether or not the host is awarded a *superhost* and some other features. Making use of [some assumptions](https://github.com/BrittaInData/dsc-5-capstone-project-online-ds-ft-100118/blob/master/02_Notebook_2.ipynb), I calculated an estimate for each accommodation's occupancy and income which we now want to visualize.

***Handling Missing Values***

```python
print('Proportion of Missing Values in Airbnb Table:', 
      round(airbnb.isna().sum().sum()/len(airbnb),3), '%')
```

```python
airbnb = airbnb.dropna()
```

```python
print('Proportion of Missing Values in Airbnb Table AFTER cleaning:', 
      round(airbnb.isna().sum().sum()/len(airbnb),3), '%')
```

***Features of Interest:***

Imagine, we wanted to know which districts are the most in-demand areas. To detect this, we just need to group by neighbourhood (or: district) and compare the estimated `modest_occupancy` and/or `optimistic_occupancy`. Again, if you want to know more about the ideas behind these concepts and how it is calculated, check my [original notebook](https://github.com/BrittaInData/dsc-5-capstone-project-online-ds-ft-100118/blob/master/02_Notebook_2.ipynb).

***A. Basic Lollipop Chart***

First we prepare the data exactly the way we need it and create a new dataframe used only for plotting:

```python
# create new dataframe by subsetting the columns we need, grouping by neighbourhood ...
# and calculating the average occupancy (optimistic estimate) for each neighbourhood
df_plot = airbnb[['neighbourhood', 'optimistic_occupancy']].groupby('neighbourhood').apply(lambda x: x.mean())
df_plot = df_plot.sort_values('optimistic_occupancy')
df_plot = df_plot.reset_index()
```

```python
# set style
sns.set_style('white')

# prepare figure
fig, ax = plt.subplots(figsize=(8,8))

# draw horizontal lines
ax.hlines(y=df_plot.index, xmin=0, xmax=df_plot['optimistic_occupancy'], 
          color='lightsteelblue', alpha=0.5, linewidth=2)

# draw dots
ax.scatter(y=df_plot.index, x=df_plot['optimistic_occupancy'], s=75, 
           color='navy', alpha=1, label='Optimistic Occupancy Estimate')

# add legend
ax.legend()

# add title
ax.set_title('\nAverage Occupancy Estimate by District\n', fontdict={'size':12, 'weight':'bold'})

# define xlabel
ax.set_xlabel('\nOccupancy Estimate')

# define yticks and ylabels
ax.set_yticks(df_plot.index)
ax.set_yticklabels(df_plot['neighbourhood'], rotation=0, fontdict={'horizontalalignment': 'right', 'size':12})

# remove frame around plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# put legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

plt.show()
```

***B. Variation: Lollipop Chart with Highlight***

```python
# highlight specific district: Mitte
my_color = np.where(df_plot['neighbourhood'] == 'Mitte', 'darkorange', 'silver')
my_size  = np.where(df_plot['neighbourhood'] == 'Mitte', 70, 30)
```

```python
# set style
sns.set_style('white')

# prepare figure
fig, ax = plt.subplots(figsize=(8,8))

# draw horizontal lines while referring to defined color and size above (highlight!)
ax.hlines(y=df_plot.index, xmin=0, xmax=df_plot['optimistic_occupancy'], 
          color=my_color, alpha=0.5, linewidth=2)

# draw dots while referring to defined color and size above (highlight!)
ax.scatter(y=df_plot.index, x=df_plot['optimistic_occupancy'], s=my_size, 
           color=my_color, alpha=1, label='Optimistic Occupancy Estimate')

# add legend
ax.legend()

# add title
ax.set_title('\nAverage Occupancy Estimate by District\n', fontdict={'size':12, 'weight':'bold'})

# define xlabel
ax.set_xlabel('\nOccupancy Estimate')

# define yticks and ylabels
ax.set_yticks(df_plot.index)
ax.set_yticklabels(df_plot['neighbourhood'], rotation=0, fontdict={'horizontalalignment': 'right', 'size':12})

# remove frame around plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# put legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

plt.show()
```

***C. Variation: Connected Dotplot (or: Dumbbell Plot, Cleveland Dotplot ...)***

This comes handy when we want to compare the value of 2 numeric values for each group. It could also be visualized using a grouped or stack barplot. However, this representation is less cluttered and way easier to read. So use it if you have 2 subgroups per group:

```python
# create new dataframe by subsetting the columns we need, grouping by neighbourhood ...
# and calculating the average occupancy (optimistic AND modest estimate) for each neighbourhood
df_plot = airbnb[['neighbourhood', 'modest_occupancy', 'optimistic_occupancy']]\
                        .groupby('neighbourhood')\
                        .apply(lambda x: x.mean())

df_plot.sort_values('modest_occupancy', inplace=True)
df_plot.reset_index(inplace=True)
```

```python
# set style
sns.set_style('white')

# prepare figure
fig, ax = plt.subplots(figsize=(8,8))

# draw horizontal lines limited by both subgroups
ax.hlines(y=df_plot.index, xmin=df_plot['modest_occupancy'], xmax=df_plot['optimistic_occupancy'], 
          color='lightgrey', alpha=0.7, linewidth=2)

# draw dots on the left and on the right
ax.scatter(y=df_plot.index, x=df_plot['modest_occupancy'],  s=75, color='skyblue', 
           alpha=0.7, label='Modest Occupancy Estimate')
ax.scatter(y=df_plot.index, x=df_plot['optimistic_occupancy'], s=75, color='darkblue', 
           alpha=0.7, label='Optimistic Occupancy Estimate')

# add legend
ax.legend()

# add title
ax.set_title('\nAverage Occupancy Estimate in % by District\n', fontdict={'size':12, 'weight':'bold'})

# define xlabel
ax.set_xlabel('\nOccupancy Estimate')

# define yticks and ylabels
ax.set_yticks(df_plot.index)
ax.set_yticklabels(df_plot['neighbourhood'], rotation=0, fontdict={'horizontalalignment': 'right', 'size':12})

# remove frame around plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# put legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

plt.show()
```
